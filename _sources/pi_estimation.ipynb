# Application of Basel Problem

#### One application of this famous formula is to estimate the value of $\pi$ by brutal force. To do so, first we carry out the following calculation

Recall formula of the problem can be expressed as the follow

$$
\sum_{n=1}^{\infty} \frac{1}{x^2} = \frac{1}{1^2} + \frac{1}{2^2} + \frac{1}{3^2} + \cdots = \frac {\pi^2}{6}
$$

With algebra, we can isolate $\pi$

$$
\pi = \sqrt[]{6\times \left( \sum_{n=1}^{\infty} \frac{1}{x^2}\right)}
$$

```python
#Import numpy
import numpy as np
```

Define the function $f(x) = \frac{1}{x^2}$

```python
def f(x):
    return 1/x**2
```

Using a for loop to calculate the Basel Problem with brutal force

```python
summation = 0

for x in np.arange(1,10000):
    result = f(x)
    summation += result

pi_estimation = (summation*6)**0.5

pi_estimation
```

```python
margin_of_error = np.pi - pi_estimation
```

```python
print(margin_of_error)
```

As we can see, the margin of error is reasonably small. Of course, this method is far from optimal
